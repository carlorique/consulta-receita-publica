```python
#Consulta de Receita Pública | Portal da Transparência do Governo Federal
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

caminho_arquivo = '/content/drive/MyDrive/ml/receitaOrcamentaria.csv'

df_orc = pd.read_csv(caminho_arquivo, encoding="utf-8-sig", low_memory=False, sep=";")

```

##Distribuição e categorias

```python
#GRÁFICO DE PIZZA
plt.figure(figsize=(6, 6))
df_orc['Categoria'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribuição de categorias')
df_orc['Categoria'] = df_orc['Categoria'].str.extract(r'\d+\s-\s(.*)')
plt.ylabel('')
plt.show()
```

##Gráfico de barras horizontais mostrando os 10 maiores valores orçamentais de 'Valor Atualizado' por órgão superior

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_orc, y='Órgão Superior', order=df_orc['Órgão Superior'].value_counts().index[:10])
plt.title('Órgãos Superiores que geraram maior receita')
plt.xlabel('Contagem')
plt.ylabel('Órgão Superior')
df_orc['Órgão Superior'] = df_orc['Órgão Superior'].str.extract(r'\d+\s-\s(.*)')
plt.show()
```

##Gráfico de Linha da Contagem de Ocorrências de origem das receitas

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_orc, x='Origem', order=df_orc['Origem'].value_counts().index[:7])
plt.title('Contagem de Ocorrências por Origem (Top 7)')
df_orc['Origem'] = df_orc['Origem'].str.extract(r'\d+\s-\s(.*)')
plt.xlabel('Origem')
plt.ylabel('Ocorrência')
plt.xticks(rotation=45)
plt.show()

```

##Gráfico de Linha da Contagem de Ocorrências ao longo das Espécies

```python
plt.figure(figsize=(10, 6))
df_orc.groupby('Espécie').size().plot(kind='line')
plt.title('Contagem de Ocorrências ao longo das Espécies')
df_orc['Espécie'] = df_orc['Espécie'].str.extract(r'\d+\s-\s(.*)')
plt.xlabel('Espécie')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

##Gráficos de barras da contagem por categoria

```python
from matplotlib import pyplot as plt
import seaborn as sns
df_orc.groupby('Categoria').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
df_orc['Unidade Gestora'] = df_orc['Unidade Gestora'].str.extract(r'\d+\s-\s(.*)')

plt.gca().spines[['top', 'right',]].set_visible(False)
plt.title('Contagem de Categorias')
```

```python
import matplotlib.pyplot as plt


df_orc['Valor Atualizado'] = df_orc['Valor Atualizado'].replace('[\.,]', '', regex=True).astype(float).round(9)

comparacao_unidades = df_orc.groupby('Unidade Gestora')['Valor Atualizado'].sum().reset_index()

comparacao_unidades = comparacao_unidades.sort_values(by='Valor Atualizado', ascending=False)

comparacao_unidades.reset_index(drop=True, inplace=True)

top_10_unidades = comparacao_unidades.head(10)

max_chars = 20
top_10_unidades['Unidade Gestora'] = top_10_unidades['Unidade Gestora'].apply(lambda x: x[:max_chars] + '...' if len(x) > max_chars else x)

plt.figure(figsize=(12, 6))
plt.bar(top_10_unidades['Unidade Gestora'], top_10_unidades['Valor Atualizado'], color='skyblue')
plt.xlabel('Unidade Gestora')
plt.ylabel('Valor')
df_orc['Unidade Gestora'] = df_orc['Unidade Gestora'].str.extract(r'\d+\s-\s(.*)')
plt.title('As dez Unidades Gestoras com Maior Valor Atualizado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

##Soma total da categoria

```python

df_orc['Valor LOA'] = df_orc['Valor LOA'].replace('[\.,]', '', regex=True).astype(float)
df_orc['Valor Atualizado'] = df_orc['Valor Atualizado'].replace('[\.,]', '', regex=True).astype(float)
comparacao_valores = df_orc.groupby('Ano').agg({'Valor LOA': 'sum', 'Valor Atualizado': 'sum'}).reset_index()
comparacao_valores['Diferenca'] = comparacao_valores['Valor Atualizado'] - comparacao_valores['Valor LOA']
plt.figure(figsize=(5, 4))
bar_width = 0.35
bar_positions1 = comparacao_valores['Ano']
bar_positions2 = [pos + bar_width for pos in bar_positions1]

plt.bar(bar_positions1, comparacao_valores['Valor LOA'], width=bar_width, label='Valor LOA', color='skyblue')
plt.bar(bar_positions2, comparacao_valores['Valor Atualizado'], width=bar_width, label='Valor Atualizado', color='orange')
plt.bar(bar_positions2, comparacao_valores['Diferenca'], width=bar_width, label='Diferença', color='red', alpha=0.7)

plt.xlabel('Ano')
plt.ylabel('Valores')
plt.title('Comparação entre Valor LOA e Valor Atualizado')
plt.xticks(bar_positions1 + bar_width / 2, comparacao_valores['Ano'])
plt.legend()
plt.tight_layout()
plt.show()
```
